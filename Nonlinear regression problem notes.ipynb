We're gonna walk through the steps to do nonlinear regression using python to find the optimum values of $k$ and $\alpha$. First we set up our problem, entering the $t$, $C_A$ data and our initial guesses for $k$ and $\alpha$.

```python
import numpy as np
t = np.array([0,5,9,15,22,30,40,60], dtype = float)
Ca = np.array([2,1.6,1.35,1.1,0.87,0.7,0.53,0.35])
initial_guess = [0.1, 2]
```

Most of the code above is python jargon (*not very important*). 
Next, we're going to define a function that calculates the difference between our actual time $t$ and the calculated time (based on our guess for $k$ and $\alpha$).

```python
def residuals(p, t, Ca):
    k, alpha = p
    err = np.absolute(t - ((Ca[0]**(1-alpha)-Ca**(1-alpha))/(k*(1-alpha))))
    return err
```

The above function (**residuals**) takes in a guess for $k$ and $\alpha$ (in the form of a vector p), and our raw data for $t$ and $C_A$, and returns the difference between $t$ and the calculated time. Refer to Eq 5-21 in your textbook for the expression for this calculated time.

Now we do the real computation, where we try to find $t$ and $\alpha$ values that minimize this difference. Python does this using a built-in module called **leastsq**. **leastsq** takes in the function that calculates the error, an initial guess, and the dependent and independent variables. 

In our case, the function that calculates the error is **residuals**, our initial guess has been defined above, our dependent variable is $t$ and our independent variable is $C_A$.     

```python
from scipy.optimize import leastsq
final_value = leastsq(residuals, initial_guess, args = (t, Ca))
k, alpha = final_value[0]
print "k = ", k, "alpha = ", alpha
```

**leastsq** returns the optimum value for our parameters, $k$ and $\alpha$, along with some other stuff such as the residual value (defined in Eq 5-22 in your textbook), and how confident it is in the $k$ and $\alpha$ values it calculated.

Before we plot the data and our regression model to see how accurate the fit is, we need to calculate some $t$ values at different $C_A$ for the calculated $k$ and $\alpha$

```python
# Making Ca and t for plotting
Ca_lsq = np.linspace(Ca[0], Ca[-1], 100)
t_lsq = ((Ca_lsq[0]**(1-alpha)-Ca_lsq**(1-alpha))/(k*(1-alpha)))
```

```python
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
import matplotlib.pyplot as plt
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
line1, = ax1.plot(t, Ca, 'o', label = 'Raw data')
line2, = ax1.plot(t_lsq, Ca_lsq, 'r-', label = 'Regression model')
ax1.set_title('Ca vs t')
ax1.set_xlabel('t')
ax1.set_ylabel('Ca')
ax1.legend()
ax1.set_xbound(lower = -5, upper = 70)
ax1.set_ybound(lower = 0, upper = 2.5)
plt.draw()
```
